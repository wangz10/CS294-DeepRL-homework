## Examine all the mojoco envs

```python
envnames = ['Hopper-v2', 'Ant-v2', 'HalfCheetah-v2', 'Humanoid-v2', 'Reacher-v2', 'Walker2d-v2']
```

```python
def describe_env(env):
    print('Observations:', env.observation_space.dtype, env.observation_space.shape)
    print('Actions:', env.action_space.dtype, env.action_space.shape)
    print('Action space high:', env.action_space.high)
    print('Action space low:', env.action_space.low)
    print('Time step limit:', env.spec.timestep_limit)
    return
```

```python
for envname in envnames:
    env = gym.make(envname)
    print(envname)
    describe_env(env)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df_full = []
for fn in os.listdir('results/'):
    envname = fn.split('-')[0]
    df = pd.read_csv(os.path.join('results', fn))
    df['envname'] = envname
    df_full.append(df)

df_full = pd.concat(df_full)
print(df_full.shape)
```

```python
df_full.head()
```

```python
df_full = df_full.melt(id_vars='envname', value_vars=['expert_returns', 'bc_returns'],
                      var_name='policy', value_name='returns')
print(df_full.shape)
df_full.head()
```
